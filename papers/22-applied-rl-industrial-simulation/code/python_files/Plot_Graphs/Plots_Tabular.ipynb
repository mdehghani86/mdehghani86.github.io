```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = pickle.load(open("../Results/Tabular/eps_array.pickle", 'rb'))
```

```python
plt.plot(eps, label="epsilon")
plt.legend()
plt.savefig("../plots/Tabular/epsilon_decay.svg", dpi=150)
plt.show()
```

```python
cycletime_array = pickle.load(open("../Results/Tabular/cycletime_array.pickle", 'rb'))
```

```python
plt.plot(cycletime_array, label="Cycle Time")
plt.legend()
plt.savefig("../plots/Tabular/CycleTime.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_cycletime = np.array(pd.Series(np.array(cycletime_array)).rolling(window=resolution).mean() )

plt.plot(cumsum_cycletime, label="Cycle Time")
plt.legend()
plt.savefig("../plots/Tabular/CycleTime_trend.svg", dpi=150)
plt.show()
```

```python
scores = pickle.load(open("../Results/Tabular/scores_array.pickle", 'rb'))
```

```python
plt.plot(scores, label="rewards")
plt.legend()
plt.savefig("../plots/Tabular/rewards.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_rewards = np.array(pd.Series(np.array(scores)).rolling(window=resolution).mean() )

plt.plot(cumsum_rewards, label="rewards")
plt.legend()
plt.savefig("../plots/Tabular/reward_trend.svg", dpi=150)
plt.show()
```

```python
losses = pickle.load(open("../Results/Tabular/loss_array.pickle", 'rb'))
loss = np.array(losses)

mean_loss = np.mean(loss, axis=1)
```

```python
plt.plot(mean_loss, label="Loss")
plt.legend()
plt.savefig("../plots/Tabular/loss.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_loss = np.array(pd.Series(np.array(mean_loss)).rolling(window=resolution).mean() )

plt.plot(cumsum_loss, label="Loss")
plt.legend()
plt.savefig("../plots/Tabular/loss_trend.svg", dpi=150)
plt.show()
```
