```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
eps = pickle.load(open("../Results/DDQN/eps_array.pickle", 'rb'))
```

```python
plt.plot(eps, label="epsilon")
plt.legend()
plt.savefig("../plots/DDQN/epsilon_decay.svg", dpi=150)
plt.show()
```

```python
cycletime_array = pickle.load(open("../Results/DDQN/cycletime_array.pickle", 'rb'))

test_cycletimes = pickle.load(open("../Results/DDQN/testcycletime_array.pickle", 'rb'))
test_cycletime = mean(test_cycletimes)
```

```python
testcycletime_array = [test_cycletime] * len(cycletime_array)

plt.plot(cycletime_array, label="Cycle Time")
plt.plot(testcycletime_array, label="Trained Agent")
plt.legend()
plt.savefig("../plots/DDQN/CycleTime.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_cycletime = np.array(pd.Series(np.array(cycletime_array)).rolling(window=resolution).mean() )

plt.plot(cumsum_cycletime, label="Cycle Time")
plt.plot(testcycletime_array, label="Trained Agent")
plt.legend()
plt.savefig("../plots/DDQN/CycleTime_trend.svg", dpi=150)
plt.show()
```
