```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = pickle.load(open("../Results/ActorCritic/eps_array.pickle", 'rb'))
```

```python
plt.plot(eps, label="epsilon")
plt.legend()
plt.savefig("../plots/ActorCritic/epsilon_decay.svg", dpi=150)
plt.show()
```

```python
cycletime_array = pickle.load(open("../Results/ActorCritic/cycletime_array.pickle", 'rb'))
```

```python
plt.plot(cycletime_array, label="Cycle Time")
plt.legend()
plt.savefig("../plots/ActorCritic/CycleTime.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_cycletime = np.array(pd.Series(np.array(cycletime_array)).rolling(window=resolution).mean() )

plt.plot(cumsum_cycletime, label="Cycle Time")
plt.legend()
plt.savefig("../plots/ActorCritic/CycleTime_trend.svg", dpi=150)
plt.show()
```

```python
scores = pickle.load(open("../Results/ActorCritic/scores_array.pickle", 'rb'))
```

```python
plt.plot(scores, label="rewards")
plt.legend()
plt.savefig("../plots/ActorCritic/rewards.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_rewards = np.array(pd.Series(np.array(scores)).rolling(window=resolution).mean() )

plt.plot(cumsum_rewards, label="rewards")
plt.legend()
plt.savefig("../plots/ActorCritic/reward_trend.svg", dpi=150)
plt.show()
```

```python
losses = pickle.load(open("../Results/ActorCritic/loss_array.pickle", 'rb'))
loss = np.array(losses)

loss = np.mean(loss, axis=1)

actor_loss = loss[:,0]
critic_loss = loss[:,1]
```

```python
plt.plot(actor_loss, label="Actor Loss")
plt.legend()
plt.savefig("../plots/ActorCritic/Actorloss.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_loss = np.array(pd.Series(np.array(actor_loss)).rolling(window=resolution).mean() )

plt.plot(cumsum_loss, label="Actor Loss")
plt.legend()
plt.savefig("../plots/ActorCritic/Actorloss_trend.svg", dpi=150)
plt.show()
```

```python
plt.plot(critic_loss, label="Critic Loss")
plt.legend()
plt.savefig("../plots/ActorCritic/Criticloss.svg", dpi=150)
plt.show()
```

```python
resolution = 10

cumsum_loss = np.array(pd.Series(np.array(critic_loss)).rolling(window=resolution).mean() )

plt.plot(cumsum_loss, label="Critic Loss")
plt.legend()
plt.savefig("../plots/ActorCritic/Criticloss_trend.svg", dpi=150)
plt.show()
```
